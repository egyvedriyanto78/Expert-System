## EXPERT SYSTEM PROGRAM

```python
def diagnose_gejala(gejala, penyakit):
    # Additional
    if all(gejala.values()):
        return "Sekarat, segera pergi ke Rumah Sakit terdekat!"

    # Aturan 1
    elif gejala['A'] and gejala['B'] and gejala['C']:
        return penyakit['W']

    # Aturan 2
    elif gejala['D'] and gejala['A'] and gejala['E'] and gejala['F'] and gejala['G'] and gejala['H'] and gejala['I']:
        return penyakit['X']

    # Aturan 3
    elif gejala['J'] and gejala['G'] and gejala['F'] and gejala['K']:
        return penyakit['Y']

    # Aturan 4
    elif gejala['D'] and gejala['A'] and gejala['L'] and gejala['M'] and gejala['N'] and gejala['O'] and gejala['P'] and gejala['Q']:
        return penyakit['Z']

    else :
        return "Tidak Diketahui"
```

```python
# Daftar Penyakit
penyakit = {
    'W': "Penyakit Otitis Eksterna",
    'X': "Sinusitis",
    'Y': "Renitis Non – Alergika",
    'Z': "Farangitis (Radang Tenggorokan)"
}
```

```python
# Input gejala dari pengguna
gejala = {
    'A': input("Apakah Anda mengalami Sakit kepala? (ya/tidak): ") == 'ya',
    'B': input("Apakah Anda mengalami Keluar cairan? (ya/tidak): ") == 'ya',
    'C': input("Apakah Anda mengalami Tanda-tanda radang di liang telinga? (ya/tidak): ") == 'ya',
    'D': input("Apakah Anda mengalami Demam? (ya/tidak): ") == 'ya',
    'E': input("Apakah Anda mengalami PUS dan di meatus media? (ya/tidak): ") == 'ya',
    'F': input("Apakah Anda mengalami Hidung tersumbat? (ya/tidak): ") == 'ya',
    'G': input("Apakah Anda mengalami Hidung meler? (ya/tidak): ") == 'ya',
    'H': input("Apakah Anda mengalami Nyeri pipi dibawah mata? (ya/tidak): ") == 'ya',
    'I': input("Apakah Anda mengalami Selaput lendir merah dan bengkak? (ya/tidak): ") == 'ya',
    'J': input("Apakah Anda mengalami Bersin-bersin? (ya/tidak): ") == 'ya',
    'K': input("Apakah Anda mengalami Lendir di tenggorokan? (ya/tidak): ") == 'ya',
    'L': input("Apakah Anda mengalami Nyeri saat berbicara atau menelan? (ya/tidak): ") == 'ya',
    'M': input("Apakah Anda mengalami Sakit pada telinga? (ya/tidak): ") == 'ya',
    'N': input("Apakah Anda mengalami Pembengkakan kelenjar getah bening? (ya/tidak): ") == 'ya',
    'O': input("Apakah Anda mengalami Tenggorokan gatal? (ya/tidak): ") == 'ya',
    'P': input("Apakah Anda mengalami Adanya tonsil yang membengkak? (ya/tidak): ") == 'ya',
    'Q': input("Apakah Anda mengalami Suara serak? (ya/tidak): ") == 'ya'
}

# Output program
print()
hasil_diagnosis = diagnose_gejala(gejala, penyakit)
print(f"Hasil Diagnosis: {hasil_diagnosis}")
```

## ALUR KERJA PROGRAM



*   Pengguna menjawab pertanyaan-pertanyaan mengenai gejala kesehatan.
*   Input pengguna diubah menjadi dictionary ‘gejala’.
*   Program memanggil fungsi ‘diagnose_gejala’ dengan ‘gejala’ dan ‘penyakit’.
*   Fungsi melakukan pengecekan aturan berdasarkan gejala yang diinput.
*   Jika ada aturan yang sesuai, program mengembalikan diagnosis penyakit yang sesuai.
*   Hasil diagnosis ditampilkan kepada pengguna.
